# Combination of all Models trained on SWISSIMAGE on a Predict.tif

```python
import rasterio
from rasterio.merge import merge
from glob import glob
import os
from tqdm import tqdm


# Path to the folder with the rasters
ordner = r"C:\Users\claud\Downloads\merged3"

# Find all TIFs in the folder
raster_files = glob(os.path.join(ordner, "*.tif"))

print(f"{len(raster_files)} Rasterdateien gefunden.")

rasters = []
for f in tqdm(raster_files, desc="√ñffne Raster"):
    rasters.append(rasterio.open(f))

# Merge with progress bar simulation
# (merge() itself has no tqdm, but we can make the loading visible)
print("Starte Merge...")
mosaic, out_transform = merge(rasters)

# Metadaten √ºbernehmen
out_meta = rasters[0].meta.copy()
out_meta.update({
    "driver": "GTiff",
    "height": mosaic.shape[1],
    "width": mosaic.shape[2],
    "transform": out_transform
})

# Speichern mit Fortschrittsbalken (Bandweise schreiben)
with rasterio.open("merged.tif", "w", **out_meta) as dest:
    for i in tqdm(range(1, mosaic.shape[0] + 1), desc="Speichere B√§nder"):
        dest.write(mosaic[i-1], i)

print("Fertig! Ausgabe: merged.tif")
```
